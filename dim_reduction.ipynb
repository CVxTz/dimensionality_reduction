```python
import plotly.offline as offline
import plotly.graph_objs as go
import plotly
import mnist_reader
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
X, y = mnist_reader.load_mnist('data/fashion', kind='t10k')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
X_test = X_test/255.0
```

```python
mapping = {0:"T-shirt/top",
1:"Trouser",
2:"Pullover",
3:"Dress",
4:"Coat",
5:"Sandal",
6:"Shirt",
7:"Sneaker",
8:"Bag",
9:"Ankle boot"}
```

```python
#TSNE :
```

```python
#dim = 2
```

```python
X_embedded_tsne2 = TSNE(n_components=2).fit_transform(X_test)
```

```python
trace = go.Scatter(
    x = X_embedded_tsne2[:, 0],
    y = X_embedded_tsne2[:, 1],
    name = 'Fashion',
    mode = 'markers',
    text = [mapping[i] for i in y_test],
    marker = dict(
        size = 10,
        color = 10*y_test,
        
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)
data = [trace]

layout = dict(title = 'Fashion mnist',
              yaxis = dict(zeroline = False),
              xaxis = dict(zeroline = False)
             )

fig = dict(data=data, layout=layout)
plotly.offline.plot(fig, filename='styled-scatter', include_plotlyjs=False, output_type='div')
```

```python
X_embedded_tsne3 = TSNE(n_components=3).fit_transform(X_test)
```

```python
trace = go.Scatter3d(
    x = X_embedded_tsne3[:, 0],
    y = X_embedded_tsne3[:, 1],
    z = X_embedded_tsne3[:, 2],
    name = 'Fashion',
    mode = 'markers',
    text = [mapping[i] for i in y_test],
    marker = dict(
        size = 10,
        color = 10*y_test,
        
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)
data = [trace]

layout = dict(title = 'Fashion mnist',
              yaxis = dict(zeroline = False),
              xaxis = dict(zeroline = False)
             )

fig = dict(data=data, layout=layout)
plotly.offline.plot(fig, filename='styled-scatter', include_plotlyjs=False, output_type='div')
```
